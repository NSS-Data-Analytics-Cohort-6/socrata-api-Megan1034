## Make an API request that returns the months where "fire" was searched in 2016. Which month had the most searches?

```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json?year=2016&query_text=fire'

response = requests.get(endpoint) 
```

```python
response
```

```python
response.text
```

```python
searches = response.json()
searches
```

### At first it would only pull the 2014 data? i needed to get all of the years so I found where to filter data on the Socrata website.

ANSWER 1: August 2016 with 47 searches for fire

```python
endpoint = 'https://data.nashville.gov/resource/fuaa-r5cm.json?$where=query_count>100&$limit=44000'

response = requests.get(endpoint) 
```
